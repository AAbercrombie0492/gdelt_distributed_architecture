# Design Principles

1. Iterate everything: tables, charts, reports, predictions.
2. Ship intermediate output.
    - Good systems are self documenting.
    - Share incomplete assets.
    - Commit all work to git
3. Prototype experiments over implementing tasks.
4. Integrate tyranical opition of data in product management.
5. Traverse data-value pyramid as we work.
    - Level 1: Plumbing
    - Level 2: Reports
    - Level 3: Predictions
6. Discover the critical path to a killer product.
7. Get meta. Describe the process, not just the end-state.
